```python
import pandas as pd 
import os
import numpy as np
import matplotlib.pyplot as plt
import cv2
import scipy
```

```python
dir_anno = "data/Anno/"
dir_data = "data/img_align_celeba/"
dir_cropped = "cropped/train/"
dir_cropped_test = "cropped_test/test/"
```

```python
df=pd.read_csv("data/Anno/attr_align_celeba.csv")
df.shape
```

```python
def get_annotation(fnmtxt, verbose=True):
    if verbose:
        print("_"*70)
        print(fnmtxt)
    
    df=pd.read_csv(dir_anno+fnmtxt)
    print(df.shape)
    df = df.dropna()
    if verbose:
        print(" Total number of annotations {}\n".format(df.shape))
        print(df.head())
    ## cast to integer
    for nm in df.columns:
        if nm != "image_id":
            df[nm] = pd.to_numeric(df[nm],downcast="integer")
    return(df)

attr   = get_annotation("attr_align_celeba.csv",verbose=False)
align  = get_annotation("landmarks_align_celeba.csv",verbose=False)

assert np.all(align["image_id"] == attr["image_id"])
```

```python
def plot_image(align,nrow=2):
    figsize = (20,10)
    ncol = 5
    fig = plt.figure(figsize=figsize)
    N = nrow*ncol
    for i, myid in enumerate(align["image_id"][:N]):
        image = plt.imread(dir_data + "/" + myid)
        image = image/255.0

        (_, 
         lefteye_x,    lefteye_y,
         righteye_x,   righteye_y, 
         nose_x,       nose_y,
         leftmouth_x,  leftmouth_y, 
         rightmouth_x, rightmouth_y) = align.iloc[i]


        ax  = fig.add_subplot(nrow,ncol,i+1)
        ax.imshow(image)
        ax.set_title(image.shape)
        ax.scatter(lefteye_x,    lefteye_y)
        ax.scatter(righteye_x,   righteye_y)
        ax.scatter(nose_x,       nose_y)
        ax.scatter(leftmouth_x,  leftmouth_y)
        ax.scatter(rightmouth_x, rightmouth_y)
plot_image(align)
```

```python
def image_crop_center(directory,align,i):
    imgid=align['image_id'][i]
    image = plt.imread(dir_data + "/" + imgid)
    image = image/255.0
    (image_name, 
     lefteye_x,    lefteye_y,
     righteye_x,   righteye_y, 
     nose_x,       nose_y,
     leftmouth_x,  leftmouth_y, 
     rightmouth_x, rightmouth_y) = align.iloc[i]
    
    centereye_x=int((int(lefteye_x)+int(righteye_x))/2)
    centereye_y=int((int(lefteye_y)+int(righteye_y))/2)
    centermouth_x=int((int(leftmouth_x)+int(rightmouth_x))/2)
    centermouth_y=int((int(leftmouth_y)+int(rightmouth_y))/2)
    
    # SOME COMPLICATED MATH RESCALING
    ## top:mid:down=u:m:l
    ## u+m+l=1
    ## top/(top+mid+down)=u
    ## mid/(top+mid+down)=m
    ## down/(top+mid+down)=l
    ## top+mid+down=mid/m=k
    ## top=u*k
    ## down=l*k
    upper_ratio=87/223
    mid_ratio=75/223
    lower_ratio=61/223
    mid=centermouth_y-centereye_y
    
    k=mid/mid_ratio
    top=int(upper_ratio*k)
    down=int(lower_ratio*k)
    topleft_y=max(0,centereye_y-top)
    bottomright_y=min(217,centermouth_y+down)
    
    half_height=int((top+mid+down)/2)
    center_x=int((centereye_x+centermouth_x)/2)
    topleft_x=max(0,center_x-half_height)
    bottomright_x=min(177,center_x+half_height)
    
    face_crop=image[topleft_y:bottomright_y,topleft_x:bottomright_x,:]
    data = 255 * face_crop # Now scale by 255
    crop = data.astype(np.uint8)
    image_to_write = cv2.cvtColor(crop, cv2.COLOR_RGB2BGR)
    cv2.imwrite(directory+image_name, image_to_write)
```
